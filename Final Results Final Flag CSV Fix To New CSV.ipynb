```python
country_names_dict = {
    "East Timor": "Timor-Leste", 
}

def change_name(s):
    if(s in country_names_dict):
        return country_names_dict[s]
    return s

data_raw['home_team'] = data_raw['home_team'].apply(lambda s: change_name(s))
data_raw['away_team'] = data_raw['away_team'].apply(lambda s: change_name(s))
```
