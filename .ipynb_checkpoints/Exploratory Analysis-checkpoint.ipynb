```python
#Useful Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the raw data
data=pd.read_csv('results.csv')
data.head()
```

# Exploratory Analysis

```python
#Distribution of match dates (per decade)
decades=data['date'].str.slice(0,3)
decades=decades.astype(str)+'0s'
decades=decades.groupby(decades).count()
pd.DataFrame(decades).plot.bar(y=0,legend=False)
```

#### The amount of matches per year increased exponentially over time.

```python
#Distribution of the matches per team (top and low 50)
match_participants=pd.concat([data['home_team'],data['away_team']])
match_participants=match_participants.groupby(match_participants).count().sort_values(0, ascending=False)
pd.DataFrame(match_participants[:50]).plot.bar(y=0,legend=False, figsize=(20,5), title='Top 50 Teams')
pd.DataFrame(match_participants[-50:]).plot.bar(y=0,legend=False, figsize=(20,5), title='Low 50 Teams')
```

```python
#Distribution of away matches per team (top and low 50)
away_participants=data['away_team'].groupby(data['away_team']).count().sort_values(0, ascending=False)
pd.DataFrame(away_participants[:50]).plot.bar(y=0,legend=False, figsize=(20,5), title='Top 50 Teams (Away)')
pd.DataFrame(away_participants[-50:]).plot.bar(y=0,legend=False, figsize=(20,5), title='Low 50 Teams (Away)')
```

```python
#Distribution of home matches per country (top and low 50)
home_participants=data['home_team'].groupby(data['home_team']).count().sort_values(0, ascending=False)
pd.DataFrame(home_participants[:50]).plot.bar(y=0,legend=False, figsize=(20,5), title='Top 50 Teams (Home)')
pd.DataFrame(home_participants[-50:]).plot.bar(y=0,legend=False, figsize=(20,5), title='Low 50 Teams (Home)')
```

#### Teams seem to have distinct policies when it comes to away and home encounters.

```python
#Distribution of the final scores 
final_scores=pd.concat([data['home_score'],data['away_score']])
final_score_count=final_scores.groupby(final_scores).count()
pd.DataFrame(final_score_count).plot.bar(y=0,legend=False)
```

```python
#Score general stats
print('The mean score is:',np.mean(final_scores))
print('The median score is:',np.median(final_scores))
print('The standard deviation of the scores is:',np.std(final_scores))
```

##### The likelihood of each score decreases exponentially with the number of goals.

```python
#Distribution of the matches per tournament (top 10)
tournament_count=data['tournament'].groupby(data['tournament']).count().sort_values(0, ascending=False).drop('Friendly')
pd.DataFrame(tournament_count[:10]).plot.bar(y=0,legend=False)
```

```python
#Repartition of friendly vs competitive matches
f_vs_c=data['tournament'].copy()
f_vs_c[f_vs_c!='Friendly']='Competitive'
pd.DataFrame(f_vs_c).groupby(f_vs_c).count().plot.pie(y=0, legend=False)
```

#### Most competitive matches are organized by the FIFA and a surpringly large amount of mathces are friendly.

```python
#Distribution of the matches per country (location, top and low 50)
country_count=data['country'].groupby(data['country']).count().sort_values(0, ascending=False)
pd.DataFrame(country_count[:50]).plot.bar(y=0,legend=False, figsize=(20,5), title='Top 50 Countires')
pd.DataFrame(country_count[-50:]).plot.bar(y=0,legend=False, figsize=(20,5), title='Low 50 Countires')
```

```python
#Distribution of the matches per country (location, top and low 50)
city_count=data['city'].groupby(data['city']).count().sort_values(0, ascending=False)
pd.DataFrame(city_count[:50]).plot.bar(y=0,legend=False, figsize=(20,5), title='Top 50 Cities')
pd.DataFrame(city_count[-50:]).plot.bar(y=0,legend=False, figsize=(20,5), title='Low 50 Cities')
```

#### Countries and cities seem to have distinct policies when it comes to hosting games.

```python
#Repartition of neutral vs non-neutral fields
city_count=data['neutral'].groupby(data['neutral']).count()
pd.DataFrame(city_count[:10]).plot.pie(y=0,legend=False)
```

```python
#Comparative stats on the scores for neutral vs non-neutral fields
neutral_matches=data[data['neutral']==True]
non_neutral_matches=data[data['neutral']==False]
neutral_diff=np.abs(neutral_matches['home_score']-neutral_matches['away_score'])
non_neutral_diff=np.abs(non_neutral_matches['home_score']-non_neutral_matches['away_score'])
non_neutral_advantage=non_neutral_matches['home_score']-non_neutral_matches['away_score']

print('For Neutral Fields:')
print('The mean difference in score for neutral fields is:',np.mean(neutral_diff))
print('The median difference in score for neutral fields is:',np.median(neutral_diff))
print('The standard deviation of the difference in score for neutral fields is:',np.std(neutral_diff))

print('For Non-Neutral Fields:')
print('The mean difference in score for non-neutral fields is:',np.mean(non_neutral_diff))
print('The median difference in score for non-neutral fields is:',np.median(non_neutral_diff))
print('The standard deviation of the difference in score for non-neutral fields is:',np.std(non_neutral_diff))
print('The mean advantage in score for home teams is:',np.mean(non_neutral_advantage))
print('The median advantage in score for home teams is:',np.median(non_neutral_advantage))
print('The standard deviation of the advantage in score for home teams is:',np.std(non_neutral_advantage))
```

```python
#Repartition of home vs away team victories for non-neutral fields
non_neutral_victories=non_neutral_advantage[non_neutral_advantage!=0]
non_neutral_victories[non_neutral_advantage>0]='Home'
non_neutral_victories[non_neutral_advantage<0]='Away'
non_neutral_victories=non_neutral_victories.groupby(non_neutral_victories).count()
pd.DataFrame(non_neutral_victories).rename(columns={0:'Match Outcome'}).plot.pie(y=0, legend=False)
```

```python
#Comparative stats on the scores for away vs home teams
print('For Home Teams:')
print('The mean score for home teams is:',np.mean(non_neutral_matches['home_score']))
print('The median score for home teams is:',np.median(non_neutral_matches['home_score']))
print('The standard deviation of the score for home teams is:',np.std(non_neutral_matches['home_score']))

print('For Away Teams:')
print('The mean score for away teams is:',np.mean(non_neutral_matches['away_score']))
print('The median score for away teams is:',np.median(non_neutral_matches['away_score']))
print('The standard deviation of the score for home teams is:',np.std(non_neutral_matches['away_score']))
```

#### Most matches take place on a neutral field.
#### Matches on non-neutral fields are mostly won by the home team.
#### The average score of for home teams is considerably higher than the one of away teams.
#### Matches on a neutral field have more varied results and higher mean score difference.
